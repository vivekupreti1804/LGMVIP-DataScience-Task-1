```python
# Step-1. Import the Reuired Library i.e:- cv2
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Step-2. Read the image which you have to converted
image = cv2.imread("vivek.jpg")
cv2_imshow(image)
```

```python
# Step-3. Convert the Original Image to Greyscale
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

```python
# Step-4. Inverted the new Greyscale image
invert_image = 255 - grey_image
cv2_imshow(invert_image)
```

```python
# Step-5. Blur the image by using the Gaussian Function in OpenCV
blur_image = cv2.GaussianBlur(invert_image, (21, 21), 0)
```

```python
# Step-6. Final Step Invert the Blurred image
inverted_blurred = 255 - blur_image
pencil_sketch = cv2.divide(grey_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
```

```python
#Final Output :- Original And Pencil Sketch Image
cv2_imshow(image)
cv2_imshow(pencil_sketch)
```
